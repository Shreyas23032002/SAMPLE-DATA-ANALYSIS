```python
ax = sns.countplot(x='Month Number', data = df )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['  Sales '], as_index=False)[' COGS '].sum().sort_values(by=' COGS ', ascending=False).head(300)

sns.barplot(x = '  Sales ',y= ' COGS ',data = sales_gen)
```

```python
ax = sns.countplot(data = df, x ='Units Sold', hue = ' Discount Band ')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby([' Discount Band '],as_index= False)[' COGS '].sum().sort_values(by= ' COGS ', ascending = False)
sns.barplot(x=' Discount Band ',y=' COGS ',data=sales_age)
```

```python
sales_state = df.groupby([' Month Name '], as_index = False)[' Manufacturing Price '].sum().sort_values(by=' Manufacturing Price ', ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x =' Month Name ',y=' Manufacturing Price ')
```

```python
sales_state = df.groupby([' Month Name '], as_index = False)[' COGS '].sum().sort_values(by=' COGS ', ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x =' Month Name ',y=' COGS ')
```

```python
ax = sns.countplot(data= df , x='Discount')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Month Number','Discount'], as_index = False)[' COGS '].sum().sort_values(by=' COGS ', ascending= False).head(10)
sns.set(rc={'figure.figsize':(6,7)})
sns.barplot(data=sales_state, x ='Month Number',y=' COGS ',hue = 'Discount')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x= ' Product ')
for bar in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby([' Product '], as_index=False)[' COGS '].sum().sort_values(by=' COGS ', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = ' Product ',y= ' COGS ') 
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Segment')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby([' Product '], as_index=False)[ ' COGS '].sum().sort_values(by= ' COGS ', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = ' Product ',y=  ' COGS ')
```
